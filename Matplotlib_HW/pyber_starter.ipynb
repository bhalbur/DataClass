```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city = pd.read_csv("data/city_data.csv")
rides = pd.read_csv("data/ride_data.csv")
```

```python
merged = pd.merge(rides, city, how='left')
merged.head()
```

```python
trim = merged.loc[:,['city','fare','driver_count','type']]
trim.head()
```

```python
grp = trim.groupby('city').agg({'fare': ['count', 'mean']})
grp.head()
```

```python
cityclean = pd.DataFrame(
    {
     'num_rides':grp['fare','count'],
     'avg_fare':grp['fare','mean'],
    }
)
cityclean.head()
```

```python
cityind = city.set_index('city')
```

```python
cleanwtype = pd.merge(cityclean, cityind, how='left', left_index=True, right_index=True)
cleanwtype.head()
```

```python
urban = cleanwtype.loc[cleanwtype['type']=='Urban'].plot.scatter(
    x='num_rides', y='avg_fare', s=cleanwtype['driver_count']*10, alpha=0.5, color ='orange', label='Urban')
```

```python
rural = cleanwtype.loc[cleanwtype['type']=='Rural'].plot.scatter(
    x='num_rides', y='avg_fare', s=cleanwtype['driver_count']*10, alpha=0.5, color="gold", label='Rural')
```

```python
suburban = cleanwtype.loc[cleanwtype['type']=='Suburban'].plot.scatter(
    x='num_rides', y='avg_fare', s=cleanwtype['driver_count']*10, alpha=0.5, label='Suburban')
```

## Bubble Plot of Ride Sharing Data

```python
fig, ax = plt.subplots()
urban = cleanwtype.loc[cleanwtype['type']=='Urban'].plot.scatter(
    x='num_rides', y='avg_fare', s=cleanwtype['driver_count']*5, alpha=0.5, color ='orange', label='Urban', ax=ax)
rural = cleanwtype.loc[cleanwtype['type']=='Rural'].plot.scatter(
    x='num_rides', y='avg_fare', s=cleanwtype['driver_count']*5, alpha=0.5, color="gold", label='Rural', ax=ax)
suburban = cleanwtype.loc[cleanwtype['type']=='Suburban'].plot.scatter(
    x='num_rides', y='avg_fare', s=cleanwtype['driver_count']*5, alpha=0.5, label='Suburban', ax=ax)

plt.title('Rides and Fares by City')
plt.text(4, 38.2, 'Circle size indicates number of drivers')
_ = plt.show
plt.savefig('Bubble_Chart.png')
```

## Total Fares by City Type

```python
cityfares = merged[['fare','type']].groupby('type').sum()
cityfares['pct'] = cityfares['fare']/sum(cityfares['fare'])
piefares = cityfares['pct']
pie1 = plt.pie(piefares, labels=piefares.index, autopct='%.1f%%', 
               explode=[0,0,.1], startangle=-90, counterclock=False, shadow=.01)
plt.title('% of total Fare $$$ by City Type')
plt.savefig('Total Fares by City Type.png')
```

## Total Rides by City Type

```python
merged['type'].value_counts()
pie2 = plt.pie(merged['type'].value_counts(), labels=merged['type'].value_counts().index, autopct='%.1f%%', 
               explode=[.1,0,0], startangle=150, counterclock=False, shadow=.01, colors=['#2CA02C','#FF7F0E','#1F77B4'])
plt.title('% of Rides by City Type')
plt.savefig('Rides by City Type.png')
```

## Total Drivers by City Type

```python
citydrivers = city[['type','driver_count']].groupby('type').sum()
citydrivers['pct'] = citydrivers['driver_count']/sum(citydrivers['driver_count'])
piedrivers = citydrivers['pct']
citydrivers
pie3 = plt.pie(piedrivers, labels=piedrivers.index, autopct='%.1f%%', 
               explode=[0,0,.1], startangle=-90, counterclock=False, shadow=.01)
plt.title('# of Drivers by City Type')
plt.savefig('Drivers by City Type.png')
```
