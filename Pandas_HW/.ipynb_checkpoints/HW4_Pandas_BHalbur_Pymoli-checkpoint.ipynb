```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
pdata = pd.read_csv(file_path)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
pd.options.display.float_format = '${:,.2f}'.format

#this gets the a list of unique Item ID + Price to calculate the true average price of all the unique items
#otherwise you are just calculating the average price of each sale
#it is basically the same number but that's just by chance
avgpxdf = pdata[['Item ID', 'Price']].drop_duplicates(['Item ID','Price'])


avgpx = avgpxdf.Price.mean()
unitems = pdata['Item ID'].nunique()
numpurch = pdata['Purchase ID'].count()
totrev = pdata['Price'].sum()

itemdf = pd.DataFrame([
    {"Unique Items": unitems, "Average Price": avgpx, "Number of Purchases": numpurch, "Total Revenue": totrev}
])
itemdf = itemdf[['Unique Items', 'Average Price', 'Number of Purchases', 'Total Revenue']]
itemdf
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#first returns a dataframe containing only unique Screen Names
unique_SN = pdata.drop_duplicates(['SN'])
unique_SN['Gender'].value_counts()
gender_list = unique_SN['Gender'].value_counts()
gender_list

```

```python
genderdf = pd.DataFrame(gender_list)
genderdf['Percentage'] = genderdf.Gender/unique_SN.Gender.count()*100
genderdf = genderdf.rename(columns={"Gender": "Total Count"})
pd.options.display.float_format = '{:,.2f}%'.format
genderdf
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting

* Display the summary data frame

```python
pd.options.display.float_format = '${:,.2f}'.format
genstats = pdata.groupby(['Gender'])
gstats = genstats.agg(
{'Price':['sum','mean','count'],'SN': pd.Series.nunique}
)
gstats
```

```python
gstats['Avg Purch Per user'] = gstats['Price','sum']/gstats['SN','nunique']
gstats[['Price','Avg Purch Per user']]
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
agebins = [0,9,14,19,24,29,34,39,99]
agegroups = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
unique_SN['AgeGroup'] = pd.cut(unique_SN["Age"], agebins, labels=agegroups)
age_list = unique_SN['AgeGroup'].value_counts()
age_list
```

```python
agedf = pd.DataFrame(age_list)
agedf
agedf['Percentage'] = agedf.AgeGroup/unique_SN['AgeGroup'].count()*100
agedf = agedf.rename(columns={"AgeGroup": "Total Count"})
pd.options.display.float_format = '{:,.2f}%'.format
agedf
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pdata['AgeBin'] = pd.cut(pdata["Age"], agebins, labels=agegroups)
pdata

pd.options.display.float_format = '${:,.2f}'.format
agestats = pdata.groupby(['AgeBin'])
agestats = agestats.agg(
{'Price':['sum','mean','count'],'SN': pd.Series.nunique}
)
agestats
```

```python
agestats['Avg Purch Per user'] = agestats['Price','sum']/agestats['SN','nunique']
agestats[['Price','Avg Purch Per user']]
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
topspend = pdata.groupby(['SN'])
topspend = topspend.agg(
{'Price':['sum','mean','count']}
)
topspend = topspend.sort_values(by=[('Price','count')], ascending=False)
topspend.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#in this one I figured out how to put the values nicely into a new dataframe after calculating using .agg
#could apply the same formatting to all previous calculations but I won't just due to time constraints

popitem = pdata[['Item ID', 'Item Name', 'Price']]
popgroup = popitem.groupby(['Item ID','Item Name']).agg({'Price':['count','mean','sum'] })
popgroup.sort_values([('Price','count')],inplace=True, ascending=False)

poptable = pd.DataFrame(
    {
     'Purchase Count':popgroup['Price','count'],
     'Item Price':popgroup['Price','mean'],
     'Total Purchase Value':popgroup['Price','sum'],
    }
)
poptable.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profittable = pd.DataFrame(
    {
     'Purchase Count':popgroup['Price','count'],
     'Item Price':popgroup['Price','mean'],
     'Total Purchase Value':popgroup['Price','sum'],
    }
)
profittable.sort_values(["Total Purchase Value"], ascending=False, inplace=True)
profittable.head()
```
